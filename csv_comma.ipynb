```python
import pandas as pd

# Read the file and rename columns
df = pd.read_csv('./shipkey.txt', encoding='utf-8')
df.columns = ['shipkey', 'shipname']

# Clean the columns and add quotes
df['shipkey'] = df['shipkey'].str.replace('ship key :', '').apply(lambda x: f'"{x.strip()}",')
df['shipname'] = df['shipname'].str.replace('ship name :', '').apply(lambda x: f'"{x.strip()}",')

# Define the file names
shipkey_name = 'shipkey.xlsx'
shipname_name = 'shipname.xlsx'

# Create Excel writer with the correct file path
with pd.ExcelWriter(shipkey_name) as writer:
    df[['shipkey']].to_excel(writer, sheet_name='shipkey', index=False)
    
with pd.ExcelWriter(shipname_name) as writer:
    df[['shipname']].to_excel(writer, sheet_name='shipname', index=False)

# Display the cleaned dataframes with quotes
display(df[['shipkey']], df[['shipname']])
```

```python
import pandas as pd

# Attempt to read the CSV file with different encodings
encodings = ['utf-8', 'latin1', 'iso-8859-1', 'cp1252']

for encoding in encodings:
    try:
        email = pd.read_csv('./SHIP_EMAIL.csv', encoding=encoding)
        print(f"Successfully read the file with encoding: {encoding}")
        display(email)
        break
    except UnicodeDecodeError as e:
        print(f"Failed to read the file with encoding: {encoding}. Error: {e}")

# Clean and format the data
email['SHIP NAME'] = email['SHIP NAME'].apply(lambda x: f'"{x.strip()}"')
email['EMAIL'] = email['EMAIL'].apply(lambda x: f'"{x.strip()}"')

# Group by COMPANY and aggregate SHIP NAME and EMAIL into lists
grouped = email.groupby('COMPANY').agg({
    'SHIP NAME': lambda x: list(x),
    'EMAIL': lambda x: list(x)
}).reset_index()

# Display the grouped DataFrame
display(grouped)

# If you want to save the grouped data into a new CSV file
grouped.to_csv('grouped_ship_email.csv', index=False)
```
